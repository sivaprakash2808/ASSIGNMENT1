```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url="https://www.redbus.in/online-booking/ksrtc-kerala/?utm_source=rtchometile"
```

```python
response=requests.get(url)
```

```python
response
```

```python
soup=BeautifulSoup(response.text)
```

```python
soup
```

```python
route=soup.select(".route")
```

```python
route
```

```python
firstbus=soup.select(".totalRoutes:nth-child(2)")
```

```python
firstbus
```

```python
lastbus=soup.select(".totalRoutes~ .totalRoutes+ .totalRoutes")
```

```python
lastbus
```

```python
amount=soup.select(".fare")
```

```python
amount
```

```python
import pandas
```

```python
busroute=pandas.DataFrame(route,columns=["bus route"])
```

```python
busroute
```

```python
busroute["firstbus"]=firstbus
```

```python
busroute["lastbus"]=lastbus
```

```python
busroute["amount"]=amount
```

```python
busroute
```
